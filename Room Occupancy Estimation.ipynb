```python
df.select_dtypes(include=['number']).skew()
```

```python
import numpy as np
numeric_cols = df.select_dtypes(include=['number']).columns
df_clean=df.copy()
for col in numeric_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_clean[col] = np.where(df_clean[col] < lower_bound, lower_bound, df_clean[col])
    df_clean[col] = np.where(df_clean[col] > upper_bound, upper_bound, df_clean[col])
```

```python
df_clean.select_dtypes(include=['number']).skew()
```

```python
import numpy as np
skewed_cols = ['S1_Light', 'S2_Light', 'S3_Light', 'S4_Light', 'S5_CO2', 'S4_Sound']
df_clean[skewed_cols] = np.log1p(df_clean[skewed_cols])
print(df_clean[skewed_cols].skew())
```

```python
#For columns that are highly skewed (e.g., S1_Light, S2_Light, S3_Light, S4_Light), we use the IQR method to cap extreme values.
#Values below the lower bound are replaced with the lower bound, and values above the upper bound are replaced with the upper bound.
#This limits the impact of outliers on the overall distribution without completely removing those rows.
#then Log Transformation for Moderately Skewed Columns
#For columns that are moderately skewed (e.g., S5_CO2 and S4_Sound), applied a log transformation (using np.log1p) to compress the range of values.
#The log transformation helps to normalize the distribution by reducing the influence of extreme values.
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#Encode Categorical Variables
# Convert 'Date' and 'Time' into dummy variables.
df_clean_encoded = pd.get_dummies(df_clean, columns=['Date', 'Time'], drop_first=True)

#Define Features and Target
# Target variable is 'Room_Occupancy_Count'
X = df_clean_encoded.drop("Room_Occupancy_Count", axis=1)
y = df_clean_encoded["Room_Occupancy_Count"]
```

```python
#checking for class imbalance, here we can se 0.0 so class is balanced so nothing to be done
y.value_counts()
```

```python
#Feature Selection using Random Forest
# Train a RandomForestClassifier to obtain feature importances.
#This code trains a Random Forest Classifier with 100 decision trees to learn patterns from the data. 
#It fits the model using X (features) and y (target), helping identify important features for better model performance. 
#The random_state=42 ensures consistent results.
# Train a Random Forest Classifier on all features.
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X, y)

# Extract feature importances from the Random Forest model.
feature_importances = pd.Series(rf_model.feature_importances_, index=X.columns)
feature_importances = feature_importances.sort_values(ascending=False)

# Use a threshold to select features.
threshold = 0.06
selected_features = feature_importances[feature_importances > threshold].index
```

```python
#Scale the Selected Features using Standardization
scaler = StandardScaler()
#X_selected_scaled = scaler.fit_transform(X[selected_features])
X_scaled = scaler.fit_transform(X)

#####Here iam using all the featues because ,i have tried various threshold values but theres always error encountering 'ValueError: at least one array or dtype is required'
###hence iam using the whole features
#Split Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

#shapes to verify splits.
#successfully data split for training and testing and ready for model training
print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```
